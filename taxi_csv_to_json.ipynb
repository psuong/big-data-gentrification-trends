```python
import pandas as pd
```

```python
taxi_df = pd.read_csv('csv_files/taxi_merge.csv')
taxi_df
```

```python
from collections import OrderedDict
grouped_neighborhoods = OrderedDict()
```

```python
for row in taxi_df.iterrows():
    if row[1]['neighborhood'] == '':
        continue
    if not str(row[1]['pickup_year']).isdigit():
        continue
    try:
        grouped_neighborhoods[row[1]['neighborhood']].append({
                'count': row[1]['count'],
                'avg_tip': row[1]['avg_tip'],
                'year': row[1]['pickup_year']
            })
    except KeyError:
        grouped_neighborhoods[row[1]['neighborhood']] = []
        grouped_neighborhoods[row[1]['neighborhood']].append({
                'count': row[1]['count'],
                'avg_tip': row[1]['avg_tip'],
                'year': row[1]['pickup_year']
            })
```
