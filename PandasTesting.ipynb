```python
import pandas as pd
```

```python
median_housing = pd.read_csv('data/output_median.csv')
```

```python
median_housing
```

```python
from collections import OrderedDict
grouped_neighborhoods = OrderedDict()
```

```python
for row in median_housing.iterrows():
    try:
        grouped_neighborhoods[row[1]['neighborhood']].append(
            {
                'familysize': row[1]['familysize'],
                'price': row[1]['price'],
                'year': row[1]['year']
            }
        )
    except KeyError:
        grouped_neighborhoods[row[1]['neighborhood']] = []
        if '2-' in row[1]['familysize']:
            row[1]['familysize'] = row[1]['familysize'].strip('2-')
        elif '+' in row[1]['familysize']:
            row[1]['familysize'] = row[1]['familysize'].strip('+')
        elif 'condo' in row[1]['familysize']:
            row[1]['familysize'] = 20
        grouped_neighborhoods[row[1]['neighborhood']].append(
            {
                'familysize': row[1]['familysize'],
                'price': row[1]['price'],
                'year': row[1]['year']
            }
        )
```

```python
for key in grouped_neighborhoods.keys():
    if str(key) == 'nan':
        del grouped_neighborhoods[key]
```
