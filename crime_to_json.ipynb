```python
import pandas as pd
```

```python
crime_df = pd.read_csv('csv_files/crime_occurrences_merge.csv')
```

```python
crime_df
```

```python
from collections import OrderedDict
grouped_neighborhoods = OrderedDict()
```

```python
for row in crime_df.iterrows():
    if str(row[1]['neighborhood']) == '' or str(row[1]['count']) == '':
        continue
    try:
        grouped_neighborhoods[row[1]['neighborhood']].append({
                'year': row[1]['occurrence_year'],
                'count': row[1]['count']
            })
    except KeyError:
        grouped_neighborhoods[row[1]['neighborhood']] = []
        grouped_neighborhoods[row[1]['neighborhood']].append({
                'year': row[1]['occurrence_year'],
                'count': row[1]['count']
            })
```

```python
grouped_neighborhoods.keys()
```

```python
import json
with open('static/data/crime_data.json', 'w') as json_file:
    json_file.write(json.dumps(grouped_neighborhoods))
```
